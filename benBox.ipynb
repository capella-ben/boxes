# BenBox

```python
from IPython.display import SVG, display
import os
from benBox import BENBOX
```

```python
b = BENBOX()

fn = 'benBox2.svg'
b.parseArgs(['--reference=0', '--debug=0', '--output=' + fn])
b.thickness = 2.8
b.tabs = 0
b.burn = 0.095       # larger numbers for a tighter fit
b.labels = False

b.outside = True
b.heightBase = 40
b.heightLid = 20
b.x = 80               # should be longest as it will have the hinge
b.y = 40

b.hinge=True
b.hingeHoleSize = 2
b.hingeEdgeOffset = 3.9
b.hingeHoleSpacing = 11
b.lid = True
b.edge_types = 'zeee'

b.open()
b.render()
b.close()
display(SVG(fn))
#os.close(fd)
#os.remove(fn)
```
